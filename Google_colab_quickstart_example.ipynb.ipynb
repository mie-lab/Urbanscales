**Run time considerations**

* For each city, running a single scale takes approximately 1.5 hours on a single threaded computer 2.6 GHz and scales almost linearly with the number of threads. We have tested upto 20 threads for each city.

* Recurrent and Non-recurrent congestion **CANNOT** be run in parallel since there is a risk of files being over-written

* Multiple cities **CAN** be run in parallel. This can be easily achieved by running the script `bash run_multiple_cities_together.sh` inside urbanscales.

Once all the csv files are generated, all the csv files generated should be copied to the `urbanscales_results` folder. The data are extracted from the log files for plotting using the following three scripts inside `urbanscales_results` folder:
* `bash extract_fig_1.sh`
* `bash extract_fig_2.sh`
* `bash extract_fig_3.sh`

Finally, the python files inside `urbanscales_results` folder will plot all the figures for the paper.
