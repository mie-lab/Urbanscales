```python
from smartprint import smartprint as sprint
import osmnx as ox

G = ox.graph_from_point((1.2927041037141997, 103.80523200806068), 50, simplify=False)
ox.plot_graph(G)
sprint (ox.basic_stats(G))

import networkx as nx
nx.draw(G)
```

```python
# simplify=True does not even run for smaller graphs.

G = ox.graph_from_point((1.2927041037141997, 103.80523200806068), 50, simplify=True)
ox.plot_graph(G)
sprint (ox.basic_stats(G))
```

```python
from smartprint import smartprint as sprint
import osmnx as ox
import networkx as nx

print ("First, with the non-simplified version")
G = ox.graph_from_point((1.2773079675770067, 103.85616178948771), 140, simplify=False)
ox.plot_graph(G)
sprint (ox.basic_stats(G))
nx.draw(G)
```

```python
print ("Now, with the simplified version")
G = ox.graph_from_point((1.2773079675770067, 103.85616178948771), 140, simplify=True)
ox.plot_graph(G)
sprint (ox.basic_stats(G))

nx.draw(G)
```
