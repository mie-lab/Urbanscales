```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Example heatmap data
data = np.random.rand(10, 100)  # Assuming you have 100 x-axis labels

plt.figure(figsize=(15, 8))
ax = sns.heatmap(data, annot=False, cmap="viridis", cbar_kws={'label': 'Feature Importance (mean |SHAP|)'}, 
                 yticklabels=True, xticklabels=True)

# Rotate the x labels by 90 degrees and show every 10th label
ax.set_xticks(np.arange(0, data.shape[1], 10))
ax.set_xticklabels(np.arange(0, data.shape[1], 10), rotation=90)

plt.title("Heatmap with Rotated X Labels and Sparse Labels")
plt.ylabel("Feature")
plt.xlabel("City-Scale Combination")
plt.tight_layout()

# Customize the colorbar manually
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=14)  # Adjust the tick font size as needed
cbar.set_label('Feature Importance (mean |SHAP|)', size=16)  # Adjust the label font size as needed

plt.savefig("heatmap_with_rotated_labels.png", dpi=300)
plt.show()
```
